```python
import cv2
print(cv2.__version__)
import os
import shutil
import Image
import ImageStat
import math
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
import time 

import io
import base64
from IPython.display import HTML

import datetime
import imageio
```

```python
#extract the candle flame image frames

vidcap = cv2.VideoCapture('candle.mp4')
success,image = vidcap.read()
frame_count = 0
success = True

#delete any previous raw image folder and re-create
try:
    shutil.rmtree("candle_imgs")
except:
    print("dir does not exist - moving on")
os.mkdir("candle_imgs")

# https://stackoverflow.com/questions/33311153/python-extracting-and-saving-video-frames
# extract frames
while success:
  cv2.imwrite("candle_imgs/%d.jpg" % frame_count, image)     # save frame as JPEG file
  success,image = vidcap.read()
  frame_count += 1
print(frame_count)
```

```python
# https://stackoverflow.com/questions/3490727/what-are-some-methods-to-analyze-image-brightness-using-python
# Average pixels, then transform to "perceived brightness".
def perceived_brightness( im_file ):
   im = Image.open(im_file)
   stat = ImageStat.Stat(im)
   r,g,b = stat.mean
   return math.sqrt(0.241*(r**2) + 0.691*(g**2) + 0.068*(b**2))

#Convert image to greyscale, return average pixel brightness.
def avg_grayscale_brightness( im_file ):
   im = Image.open(im_file).convert('L')
   stat = ImageStat.Stat(im)
   return stat.mean[0]
```

```python
frames_perceived_brightness = []

# get avg perceived brightness of each frame
for x in range(0, frame_count):
    frame_file = "candle_imgs/%d.jpg" % (x)
    frame_perceived_brightness = round(perceived_brightness(frame_file))
    frames_perceived_brightness.append(frame_perceived_brightness)
```

```python
# get some numerical stats on the numbers
max_value = max(frames_perceived_brightness)
min_value = min(frames_perceived_brightness)
avg_value = sum(frames_perceived_brightness)/len(frames_perceived_brightness)


min_stat = "Min: %d" % (min_value)
max_stat = "Max: %d " % (max_value)
avg_stat = "Avg: %d" % (avg_value)

print(min_stat, max_stat, avg_stat)
```

```python
# visualize the frames_perceived_brightness output
x = list(range(frame_count))
y = frames_perceived_brightness
plt.plot(x, y)
plt.show()
```

```python
# let's see  what the data looks like for averaged grayscale
frames_grayscale_avg = []

# get avg perceived brightness of each frame
for x in range(0, frame_count):
    frame_file = "candle_imgs/%d.jpg" % (x)
    frame_grayscale_avg = round(avg_grayscale_brightness(frame_file))
    frames_grayscale_avg.append(frame_grayscale_avg)
```

```python
# get some numerical stats on the numbers
max_value = max(frames_grayscale_avg)
min_value = min(frames_grayscale_avg)
avg_value = sum(frames_grayscale_avg)/len(frames_grayscale_avg)

min_stat = "Min: %d" % (min_value)
max_stat = "Max: %d " % (max_value)
avg_stat = "Avg: %d" % (avg_value)

print(min_stat, max_stat, avg_stat)
```

```python
#first the grayscale one

#delete any previous image folder and re-create
try:
    shutil.rmtree("candle_gray")
except:
    print("dir does not exist - moving on")
os.mkdir("candle_gray")

for idx, img_val in enumerate(frames_perceived_brightness):
    img = np.zeros([300,300,3],dtype=np.uint8)
    img.fill(img_val) # or img[:] = 255
    img_path = "candle_gray/%d.jpg" % (idx)
    cv2.imwrite(img_path, img)
```

```python
# http://www.xavierdupre.fr/blog/2016-03-30_nojs.html
def make_video(images, img_dir, outvid, outimg=None, fps=5, size=None,
               is_color=True, format="XVID"):
    """
    Create a video from a list of images.
 
    @param      outvid      output video
    @param      images      list of images to use in the video
    @param      fps         frame per second
    @param      size        size of each frame
    @param      is_color    color
    @param      format      see http://www.fourcc.org/codecs.php
    @return                 see http://opencv-python-tutroals.readthedocs.org/en/latest/py_tutorials/py_gui/py_video_display/py_video_display.html
 
    The function relies on http://opencv-python-tutroals.readthedocs.org/en/latest/.
    By default, the video will have the size of the first image.
    It will resize every image to this size before adding them to the video.
    """
    from cv2 import VideoWriter, VideoWriter_fourcc, imread, resize
    fourcc = VideoWriter_fourcc(*format)
    vid = None
    
    # patch for older version 
    try:
        FileNotFoundError
    except NameError:
        FileNotFoundError = IOError
    
    for image in images:
        image = img_dir+"/"+image
        if not os.path.exists(image):
            raise FileNotFoundError(image)
        img = imread(image)
        if vid is None:
            if size is None:
                size = img.shape[1], img.shape[0]
            vid = VideoWriter(outvid, fourcc, float(fps), size, is_color)
        if size[0] != img.shape[1] and size[1] != img.shape[0]:
            img = resize(img, size)
        vid.write(img)
    vid.release()
    return vid
```

```python
# lets make the grayscale video and simulate the output of what an led would be with the values
images = os.listdir("candle_gray")
make_video(images, "candle_gray", "candle_gray.mp4", outimg=None, fps=5, size=None, is_color=True, format="X264")
```
